```python
from V5_DataProcesser_Class import DataProcessor
# setting the initial required parameters for DataProcessor
folder_location = 'jan_8/with_record_id/'
input_csv = folder_location + 'jan_8_input_record_id.csv'

clean_csv = folder_location +'jan_8_cleaned_data.csv'
column_names = folder_location+ "jan_8_column_names.txt"
output_csv = column_names

# initialize the class

processor = DataProcessor(input_csv, clean_csv, column_names)
print("initiated class")

processor.process_and_clean_header()
print("cleaned headers")

processor.process_redcap_csv()
print("processed redcap")

# encode data
symptom_list = ['nausea', 'pain','bloat','vomit']
ambiguous_symptoms ={
                    'pain': 'obstructive'
                    }

print("inputs are made")

encoded_data_loc = folder_location+'encoded_data.csv'
encoded_data = processor.encode_and_filter_data(symptom_list, ambiguous_symptoms)
print("encoded and filtered")

processor.save_to_csv(encoded_data, encoded_data_loc)
print(f'data is saved in {encoded_data_loc}')
```
