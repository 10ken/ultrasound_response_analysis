```python
def clean_weight(x):
    
    lb_to_kg = 0.45359237
    
    # if type is string then remove spaces
    if type(x) == str:
        
        # replace all spaces
        x = x.replace(' ','')
        
        if ('lb' in x):
            
            # index of metric
            lb_index = x.index('lb')
            
            x = x[:lb_index]
            
            # remove the metric and make numeric
            x = float(x[:lb_index])
            
            # convert to kg, round to 2 decimals
            x = round(x * lb_to_kg, 3)
            
        
        elif 'kg' in x:
            
            # index of metric
            kg_index = x.index('kg')
            
            # remove the 
            x = float(x[:kg_index])
        
        # convert back to string
        x = str(x)
    
    
    elif type(x) == int or type(x) == float:
        
        # if weight is over 200kg then metric is likely in lbs
            # convert to lbs
        if x >= 200:
            x = x * lb_to_kg
    
    elif type(x) == np.nan:
        None
        
    # return the cleaned weight value
    return x
```

```python
# before clean
red_cap['weight_kg'].iloc[[0,1,57,80]]
```

```python
# after clean and updating value, (not saved to local data)
red_cap['weight_kg'] = red_cap['weight_kg'].apply(clean_weight)

red_cap['weight_kg'].iloc[[0,1,57,80]]
```
